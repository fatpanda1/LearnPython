```python
dict_a = {'name':'小明','age':35,'id':'1001'}
print(dict_a)
type(dict_a)
```

```python
dict_a['name']
```

```python
# 添加元素
dict_a['addr'] = 'qingdao'
dict_a
```

```python
dict_a.pop('addr')
dict_a
```

```python
# 获得指定的值
dict_a.get('name')
```

```python
# 如果没有指定的key，则返回，后面内容
dict_a.get('abc','hello world')
```

```python
dict_a.get('abc') is None
```

```python
# 返回所有的key
dict_a.keys()
```

```python
# 返回所有的元素
dict_a.items()
```

```python
# 遍历字典，打印所有的键/值
for key in dict_a.keys():
    print(key)
    print(dict_a[key])
```
