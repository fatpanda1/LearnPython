```python
if x < 0:
    print('x小于0')
elif(x > 0):
    print('x大于0')
else:
    print('x等于0')
```

# 循环语句

```python
fruits = ['banana','apple','orange']
for fruit in fruits:
    print(fruit)
```

### range(起始索引，结束索引，步长)，区间左闭右开

```python
for i in range(1,10,2):
    print(i)
```

```python
total = 0
```

```python
for i in range(1,101):
    total = total + i
print(total)
```

# 其他控制语句

### 列表推导式

```python
list_a = [x ** 2 for x in range(10)]
list_a
```

```python
list_b = [x **2 for x in range(10) if x > 3]
list_b
```
