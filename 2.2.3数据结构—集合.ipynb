```python
# 集合中的元素没有重复的
a_set = {1,2,3,4,5,5,5}
a_set
```

```python
# 集合是无序的
a_set[1]
```

```python
a_set.add(10)
a_set
```

```python
a_set.remove(10)
a_set
```

```python
# 1是否包含在a_set中
{1}.issubset(a_set)
```
